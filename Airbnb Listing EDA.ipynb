```python


def handle_missing_data(df,drop=False):
    if drop==True:
        df.dropna(inplace=True)
    else:
        col=df[df.isna()].columns
        for c in col:
            if df[c].dtype == 'object':
                mode_value = df[c].mode()[0]
                df[c].fillna(mode_value, inplace=True)
            elif df[c].dtype == 'int64' or df[c].dtype == 'float64':
                mean_value = df[c].mean()
                df[c].fillna(mean_value, inplace=True)
            elif df[c].dtype == 'bool':
                mode_value = df[c].mode()[0]
                df[c].fillna(mode_value, inplace=True)
handle_missing_data(df)
df.isna().sum()
```
